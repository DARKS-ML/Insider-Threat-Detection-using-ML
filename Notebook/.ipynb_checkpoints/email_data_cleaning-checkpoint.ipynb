```python
from Modules import modules as md
from Modules import usermodules as um
import pandas as pd
```

```python
df = md.ReadDataset('r4.2/email')
```

```python
display(df.shape)
display(df.head())
display(df.tail())
```

```python
#conver to date time
um.convertToDateTime(df,'date')
```

```python
#re visualize
display(df.shape)
display(df.head())
display(df.tail())
```

```python
df1 = df.copy()
```

```python
df1['year'] = df1['date'].dt.year
```

```python
df1.year.unique()
```

```python
display(df1.shape)
display(df1.head())
display(df1.tail())
```

```python
df1.info()
```

```python
new_df = df1[df1['date'].dt.year == 2010]
```

```python
display(new_df.shape)
display(new_df.head())
display(new_df.tail())
```

```python
sample_df = new_df.sample(209945)
```

```python
display(sample_df.shape)
display(sample_df.head())
display(sample_df.tail())
```

```python
sample_df.columns.unique()
```

```python
del sample_df['cc']
del sample_df['bcc']
del sample_df['attachments']
del sample_df['content']
del sample_df['year']
del sample_df['id']
```

```python
sample_df.head()
```

```python
#export new data frame
um.exportFile(sample_df,'email')
```
